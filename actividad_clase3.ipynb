+ Crea un diccionario con información de productos (nombre del producto, precio, cantidad en stock). 
+ Utiliza el método .keys() para obtener una lista de las llaves (nombres de productos) en el diccionario. 
+ Utiliza el método .values() para obtener una lista de los valores (precios y cantidades) en el diccionario.
+ Utiliza el método .items() para obtener una lista de tuplas con pares clave-valor del diccionario. 
+ Muestra cómo utilizar el método .get() para obtener el precio de un producto específico del diccionario.

```python
productos={"Vino Tinto":{"Precio":5000,"Stock":150},
           "Vino Blanco":{"Precio":3500,"Stock":100},
           "Gaseosas":{"Precio":2500,"Stock":250},
           "Jugos":{"Precio":2000,"Stock":200},
           "Agua mineral":{"Precio":1500,"Stock":300}}
```

```python
productos.keys()
```

```python
productos.values()
```

```python
productos.items()
```

```python
items_productos = productos.items()
print("Items de productos:", items_productos)
```

```python
precio_vinotinto = productos.get("Vino Tinto").get("Precio")
print("Precio del Vino Tinto: ", precio_vinotinto,"Pesos")
```

```python
precio_vinobco = productos.get("Vino Blanco").get("Precio")
print("Precio del Vino Blanco: ", precio_vinobco,"Pesos")
```

```python
precio_gaseosas = productos.get("Gaseosas").get("Precio")
print("Precio de las Gaseosas: ", precio_gaseosas,"Pesos")
```

```python
precio_jugos = productos.get("Jugos").get("Precio")
print("Precio de los Jugos: ", precio_jugos,"Pesos")
```

```python
precio_agua = productos.get("Agua mineral").get("Precio")
print("Precio del Agua Mineral: ", precio_agua,"Pesos")
```
